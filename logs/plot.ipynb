```python
import datetime
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

# root_path = os.path.abspath(os.path.dirname(__file__))
# path = os.path.join(root_path, "./test.txt")
with open("test.txt","r") as log:
    lines = log.readlines()
    
    ### x-axis is time
    base = datetime.datetime(2018, 7, 6, 15, 29)
    n_epoch = len(lines)
    now = datetime.datetime(2018, 7, 6, 17, 45) #datetime.datetime.now()
    difference = (now - base) // n_epoch
    time = [base + difference*n for n in range(n_epoch)]
    
    ### y-axes are losses
    loss_A = [line.split("\t")[0] for line in lines]
    series_loss_A = pd.Series(loss_A).rolling(100).mean()
    loss_B = [line.split("\t")[1].rsplit("\n")[0] for line in lines]
    series_loss_B = pd.Series(loss_B).rolling(100).mean()
    
# plot
plt.plot(time, series_loss_A, label="loss_A")
plt.plot(time, series_loss_B, label="loss_B")
plt.legend()
# beautify the x-labels
plt.gcf().autofmt_xdate()
plt.show()
```
